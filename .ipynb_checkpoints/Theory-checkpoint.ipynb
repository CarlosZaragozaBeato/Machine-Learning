# Linear Regression

Linear Regression is a statistical method used for modeling the linear relathionshipbetween a dependent variable and one or more independent variables. It's a supervised learning algorithm, meaning it requires labeled training data to make predictions.

In linear regression, the goal is to find the line of best fit that can be used to predict the value of the dependent variable based on the values of the independent variables. The line of best fit is the line that minimizes the sum of squared errors between the predicted values and the true values.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
```

```python
spy_potential = np.array([40,25,40,55,78,77,89,95])
spy_iq = np.array([4,3,2,5,8,7,9,10])
```

```python
model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(spy_potential, spy_iq, test_size=.35)
```

```python
model.fit(x_train.reshape(-1,1), y_train.reshape(-1,1))
```

```python
plt.xlim(0,105)
plt.ylim(0,11)
plt.xlabel(xlabel="SPY POTENTIAL", fontsize=15.0)
plt.ylabel(ylabel="IQ", fontsize=15.0)
plt.title("Spy")
plt.plot(np.linspace(0,100), model.predict(np.linspace(0,100).reshape(-1,1)), "r")
plt.scatter(spy_potential.reshape(-1,1), spy_iq.reshape(-1,1))
```

```python
model.score(x_test.reshape(-1,1), y_test.reshape(-1,1))
```

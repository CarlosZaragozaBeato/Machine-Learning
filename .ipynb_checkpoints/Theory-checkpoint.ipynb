## K-Nearest Neighbors (KNN)

K-Nearest Neighbors(KNN) is a simple, easy-to-implement, and lazy learning classification algorithm. It's a supervised learning algorithm, meaning it requires labeled training datat to make predictions.

In KNN, an object is classified by a majority vote of its neighbors, with the object being assigned to the class most common among its k nearest neighbors(k is a positive integer, typically small). If k =1, the the object is simply assigned to the class of its nearest neighbor.

KNN is a non-parametric algorithm, meaning it does not make any assumptions about the underlying data. This is useful for data that is not linearly separable.

One of the main advantages of KNN is that it's easy to implement and understand, and it's not computionally intensive. However, one of the main drawbacks is that it can be sensitive to the choice of k, and it can be slow to make predictions on large datasets.

KNN can be used for both classification and regression tasks. In classification, the goal is to predict a categorical label, such as "spam" or "not spam" for email data. In regression, the goal is to predict a continuous target variable, such as as the price of a house

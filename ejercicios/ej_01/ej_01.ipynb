```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
```

```python
data.keys()
```

```python
x = data["x"]
```

```python
x
```

```python
y = data["y"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2)
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(x_train).reshape(-1,1), np.array(y_train).reshape(-1,1))
```

```python
plt.scatter(x, y)
plt.plot(np.linspace(0,100),model.predict(np.linspace(0,100).reshape(-1,1)), "r", linewidth="3")
```

```python
model.score(np.array(x_test[:10]).reshape(-1,1), np.array(y_test[:10]).reshape(-1,1))
```

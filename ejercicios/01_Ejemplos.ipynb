```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
scores = np.array([1,4,5,6,8,9,4,5,7,8,0,1,4,2,3,4])
hours = np.array([2,10,15,20,40,50,25,5,30,40,0,2,4,1,3,15])
```

```python
model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(hours,scores, test_size=0.2)
```

```python
model.fit(x_train.reshape(-1,1), y_train.reshape(-1,1))
```

```python
plt.ylim(0,10)
plt.xlim(0,60)
plt.plot(np.linspace(0,60,100),model.predict(np.linspace(0,60,100).reshape(-1,1)), 'r')
plt.scatter(hours,scores)
```

```python
model.score(x_test.reshape(-1,1), y_test.reshape(-1,1))
```
